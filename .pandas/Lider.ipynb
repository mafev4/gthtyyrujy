```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv', encoding='utf8')
```

```python
df["n"] = df["Número de documento"].astype(str).apply(
    lambda x: sum(int(i) for i in x))
```

```python
def punto1(n):
    n1=int(n[4]) 
    c105=(n1*105**2)+(2*105)+23
    c100=(n1*100**2)+(2*100)+23
    fc1=(c105-c100)/(105-100)
    return fc1
```

```python
df["Punto 1"]=df.apply(punto1, axis=1)
```

```python
def calificación1(n):
    n1=int(n[5])
    n2=int(n[1])
    if n1==n2:
        return 5
    else:
        return 1
```

```python
df["Calificación P1"]=df.apply(calificación1, axis=1)
```

```python
df.to_csv('Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv', index=False)
```

```python
def cf(n):
    nota1=int(n[6]) 
    nota2=int(n[8]) 
    nota3=int(n[10])
    notafinal=(nota1+nota2+nota3)/3 
    return round(notafinal, 2)
```

```python
df["Calificación Final"]=df.apply(cf, axis=1)
```
