```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv', encoding='utf8')
```

```python
def c2(x):
    n=int(x[4])
    s= n*101**2+2*101+23
    s1=n*100**2+2*100+23
    sr=(s-s1)
    return float(sr)
```

```python
df["Punto 2"]=df.apply(c2,axis=1)
```

```python
def calificacion2(x):
    n=int(x[2])
    n1=int(x[7])
    if n!=n1:
        return 1
    else:
        return 5
```

```python
df["Calificacion P2"] = df.apply(calificacion2,axis=1)
```

```python
df.iloc[:, 3] = df.iloc[:, 3].astype(str).str.replace(',','.').astype(float)
```

```python
df.iloc[:, 3] = df.iloc[:, 3].astype(float)
```

```python
df['Punto 3'] = df['n'].apply(lambda x:(((x*100*2) + (2)))/(101-100))
```

```python
def calificacion3(x):
    pri = int(x[3])
    seg = int(x[9])
    if pri != seg:
        return 1
    else:
        return 5
```

```python
df['Calificacion P3'] = df.apply(calificacion3,axis = 1)
```
